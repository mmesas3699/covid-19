## Get data

```python
import csv
import requests
import pandas as pd
import io

url_data = requests.get('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/01--2020.csv').content
raw_data = pd.read_csv(io.StringIO(url_data.decode('utf-8')))
```

```python
values = raw_data.values

with open('daily.csv', 'w') as csvfile:
    for row in values:
        csvwriter = csv.writer(csvfile, delimiter=',')
        csvwriter.writerow(row)
```
