## EXTRACT data

```python
import csv
import requests
import pandas as pd
import io

url_data = requests.get('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/01-22-2020.csv').content
raw_data = pd.read_csv(io.StringIO(url_data.decode('utf-8')))
```

```python
values = raw_data.values

with open('daily.csv', 'w') as csvfile:
    for row in values:
        csvwriter = csv.writer(csvfile, delimiter=',')
        csvwriter.writerow(row)
```

```python
base_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/'
dates = ['03-17-2020', '01-23-2020', '01-24-2020']

url = base_url + dates[0] + '.csv'
url
_url = requests.get(url).content
_data = pd.read_csv(io.StringIO(_url.decode('utf-8')))
_values = _data.values
file_name = dates[0] + '.csv'
file_name
_values
```

```python

with open(file_name, 'w') as csvfile:
    for row in _values:
        csvwriter = csv.writer(csvfile, delimiter=',')
        csvwriter.writerow(row)
```
